```python
from datasets import load_dataset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

print("="*60)
print("BƯỚC 1: KHÁM PHÁ DỮ LIỆU (EDA)")
print("="*60)

# ===== 1. LOAD DATASET =====
print("\n[1/7] Đang tải dataset...")
dataset = load_dataset("uitnlp/vietnamese_students_feedback")
print("✓ Tải thành công!")

# ===== 2. XEM CẤU TRÚC =====
print("\n[2/7] Cấu trúc Dataset:")
print("-" * 60)
print(dataset)

# ===== 3. CHUYỂN SANG DATAFRAME =====
print("\n[3/7] Chuyển sang DataFrame...")
train_df = pd.DataFrame(dataset['train'])
val_df = pd.DataFrame(dataset['validation'])
test_df = pd.DataFrame(dataset['test'])

print(f"✓ Train: {len(train_df):,} mẫu")
print(f"✓ Validation: {len(val_df):,} mẫu")
print(f"✓ Test: {len(test_df):,} mẫu")
print(f"✓ Tổng cộng: {len(train_df) + len(val_df) + len(test_df):,} mẫu")

# ===== 4. XEM DỮ LIỆU MẪU =====
print("\n[4/7] 5 mẫu đầu tiên:")
print("-" * 60)
print(train_df.head())

print("\n📊 Thông tin các cột:")
print(f"  - Các cột: {train_df.columns.tolist()}")
print(f"  - Kiểu dữ liệu:")
for col in train_df.columns:
    print(f"    • {col}: {train_df[col].dtype}")

# ===== 5. KIỂM TRA MISSING VALUES =====
print("\n[5/7] Kiểm tra Missing Values:")
print("-" * 60)
missing = train_df.isnull().sum()
print(missing)
if missing.sum() == 0:
    print("✓ Không có missing values!")
else:
    print(f"⚠ Có {missing.sum()} missing values!")

# ===== 6. PHÂN TÍCH NHÃN =====
print("\n[6/7] Phân tích phân bố nhãn:")
print("-" * 60)

# Mapping nhãn
label_names = {0: 'Negative', 1: 'Neutral', 2: 'Positive'}

# Đếm sentiment
sentiment_counts = train_df['sentiment'].value_counts().sort_index()
print("\nSentiment (Cảm xúc):")
for label, count in sentiment_counts.items():
    percentage = (count / len(train_df)) * 100
    print(f"  {label} ({label_names[label]:>8}): {count:>5} mẫu ({percentage:>5.2f}%)")

# Đếm topic (nếu bạn quan tâm)
print("\nTopic (Chủ đề):")
topic_counts = train_df['topic'].value_counts().sort_index()
for topic, count in topic_counts.items():
    percentage = (count / len(train_df)) * 100
    print(f"  Topic {topic}: {count:>5} mẫu ({percentage:>5.2f}%)")

# Kiểm tra class imbalance
print("\n📊 Đánh giá Class Balance:")
max_count = sentiment_counts.max()
min_count = sentiment_counts.min()
imbalance_ratio = max_count / min_count
if imbalance_ratio > 3:
    print(f"⚠ Dataset KHÔNG cân bằng! Tỷ lệ: {imbalance_ratio:.2f}:1")
    print("  → Cần xem xét dùng class_weight hoặc resampling")
elif imbalance_ratio > 1.5:
    print(f"⚡ Dataset hơi lệch. Tỷ lệ: {imbalance_ratio:.2f}:1")
else:
    print(f"✓ Dataset cân bằng tốt! Tỷ lệ: {imbalance_ratio:.2f}:1")

# ===== 7. PHÂN TÍCH ĐỘ DÀI VÀN BẢN =====
print("\n[7/7] Phân tích độ dài văn bản:")
print("-" * 60)

# Tính độ dài
train_df['text_length'] = train_df['sentence'].apply(len)
train_df['word_count'] = train_df['sentence'].apply(lambda x: len(x.split()))

print("\nThống kê độ dài văn bản:")
print(train_df[['text_length', 'word_count']].describe())

print("\nĐộ dài trung bình theo sentiment:")
for label in sorted(train_df['sentiment'].unique()):
    subset = train_df[train_df['sentiment'] == label]
    print(f"  {label_names[label]:>8}: {subset['text_length'].mean():>6.1f} ký tự, {subset['word_count'].mean():>5.1f} từ")

# ===== 8. VẼ BIỂU ĐỒ =====
print("\n📊 Đang vẽ biểu đồ...")
fig = plt.figure(figsize=(16, 10))

colors = ['#e74c3c', '#95a5a6', '#2ecc71']  # red, gray, green

# 1. Phân bố Sentiment
ax1 = plt.subplot(2, 3, 1)
bars = ax1.bar(sentiment_counts.index, sentiment_counts.values, color=colors, edgecolor='black', linewidth=1.5)
ax1.set_title('Phân bố Sentiment', fontsize=14, fontweight='bold')
ax1.set_xlabel('Sentiment', fontsize=12)
ax1.set_ylabel('Số lượng', fontsize=12)
ax1.set_xticks(sentiment_counts.index)
ax1.set_xticklabels([label_names[i] for i in sentiment_counts.index])
ax1.grid(axis='y', alpha=0.3)
for i, (idx, val) in enumerate(sentiment_counts.items()):
    ax1.text(idx, val + 100, f'{val}\n({val/len(train_df)*100:.1f}%)', 
             ha='center', va='bottom', fontweight='bold')

# 2. Phân bố Topic
ax2 = plt.subplot(2, 3, 2)
topic_counts.plot(kind='bar', ax=ax2, color='skyblue', edgecolor='black', linewidth=1.5)
ax2.set_title('Phân bố Topic', fontsize=14, fontweight='bold')
ax2.set_xlabel('Topic', fontsize=12)
ax2.set_ylabel('Số lượng', fontsize=12)
ax2.tick_params(axis='x', rotation=0)
ax2.grid(axis='y', alpha=0.3)

# 3. Pie chart Sentiment
ax3 = plt.subplot(2, 3, 3)
ax3.pie(sentiment_counts.values, labels=[label_names[i] for i in sentiment_counts.index],
        autopct='%1.1f%%', colors=colors, startangle=90, textprops={'fontsize': 11, 'fontweight': 'bold'})
ax3.set_title('Tỷ lệ Sentiment', fontsize=14, fontweight='bold')

# 4. Boxplot độ dài theo sentiment
ax4 = plt.subplot(2, 3, 4)
sns.boxplot(data=train_df, x='sentiment', y='text_length', ax=ax4, palette=colors)
ax4.set_title('Độ dài văn bản theo Sentiment', fontsize=14, fontweight='bold')
ax4.set_xlabel('Sentiment', fontsize=12)
ax4.set_ylabel('Độ dài (ký tự)', fontsize=12)
ax4.set_xticklabels([label_names[i] for i in sorted(train_df['sentiment'].unique())])
ax4.grid(axis='y', alpha=0.3)

# 5. Boxplot số từ theo sentiment
ax5 = plt.subplot(2, 3, 5)
sns.boxplot(data=train_df, x='sentiment', y='word_count', ax=ax5, palette=colors)
ax5.set_title('Số từ theo Sentiment', fontsize=14, fontweight='bold')
ax5.set_xlabel('Sentiment', fontsize=12)
ax5.set_ylabel('Số từ', fontsize=12)
ax5.set_xticklabels([label_names[i] for i in sorted(train_df['sentiment'].unique())])
ax5.grid(axis='y', alpha=0.3)

# 6. Histogram số từ
ax6 = plt.subplot(2, 3, 6)
ax6.hist(train_df['word_count'], bins=50, color='skyblue', edgecolor='black', alpha=0.7)
mean_words = train_df['word_count'].mean()
median_words = train_df['word_count'].median()
ax6.axvline(mean_words, color='red', linestyle='--', linewidth=2, label=f'Mean: {mean_words:.1f}')
ax6.axvline(median_words, color='green', linestyle='--', linewidth=2, label=f'Median: {median_words:.1f}')
ax6.set_title('Phân bố Số từ', fontsize=14, fontweight='bold')
ax6.set_xlabel('Số từ', fontsize=12)
ax6.set_ylabel('Tần suất', fontsize=12)
ax6.legend()
ax6.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('eda_analysis.png', dpi=300, bbox_inches='tight')
print("✓ Đã lưu biểu đồ vào 'eda_analysis.png'")
plt.show()

# ===== 9. XEM VÍ DỤ TỪNG LOẠI =====
print("\n📝 Ví dụ từng loại cảm xúc:")
print("=" * 60)
for label in sorted(train_df['sentiment'].unique()):
    name = label_names[label]
    print(f"\n{name.upper()} (Sentiment {label}):")
    print("-" * 60)
    examples = train_df[train_df['sentiment'] == label]['sentence'].head(5)
    for i, text in enumerate(examples, 1):
        print(f"{i}. {text}")

# ===== 10. BẢNG TỔNG KẾT =====
print("\n📊 BẢNG TỔNG KẾT:")
print("=" * 60)
summary = train_df.groupby('sentiment').agg({
    'sentence': 'count',
    'text_length': ['mean', 'std', 'min', 'max'],
    'word_count': ['mean', 'std', 'min', 'max']
}).round(2)

summary.columns = ['Count', 'Avg_Chars', 'Std_Chars', 'Min_Chars', 'Max_Chars',
                   'Avg_Words', 'Std_Words', 'Min_Words', 'Max_Words']
summary.index = [label_names[i] for i in summary.index]
print(summary)

print("\n" + "=" * 60)
print("✅ HOÀN THÀNH BƯỚC 1: EDA")
print("=" * 60)
print("\n💡 KẾT LUẬN:")
print(f"   • Dataset có {len(train_df):,} mẫu train")
print(f"   • {len(train_df['sentiment'].unique())} loại cảm xúc: Negative, Neutral, Positive")
print(f"   • Độ dài TB: {train_df['word_count'].mean():.1f} từ/câu")
print(f"   • Class balance: {imbalance_ratio:.2f}:1")
```

Tiền Xử Lý Dữ Liệu

```python
from datasets import load_dataset
import pandas as pd
import re
import warnings
warnings.filterwarnings('ignore')

print("="*60)
print("BƯỚC 2: TIỀN XỬ LÝ DỮ LIỆU")
print("="*60)

# ===== 1. LOAD DATASET =====
print("\n[1/6] Đang tải dataset...")
dataset = load_dataset("uitnlp/vietnamese_students_feedback")
print("✓ Tải thành công!")
print(f"  Train: {len(dataset['train']):,} mẫu")
print(f"  Validation: {len(dataset['validation']):,} mẫu")
print(f"  Test: {len(dataset['test']):,} mẫu")

# ===== 2. HÀM TIỀN XỬ LÝ CƠ BẢN =====
print("\n[2/6] Định nghĩa hàm tiền xử lý...")

def preprocess_text_basic(text):
    """
    Tiền xử lý cơ bản cho văn bản tiếng Việt
    """
    if not isinstance(text, str):
        return ""
    
    # 1. Chuyển về chữ thường
    text = text.lower()
    
    # 2. Loại bỏ URL
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)
    
    # 3. Loại bỏ email
    text = re.sub(r'\S+@\S+', '', text)
    
    # 4. Loại bỏ HTML tags (nếu có)
    text = re.sub(r'<.*?>', '', text)
    
    # 5. Loại bỏ số (tùy chọn - comment dòng này nếu muốn giữ số)
    # text = re.sub(r'\d+', '', text)
    
    # 6. Loại bỏ ký tự đặc biệt, giữ lại tiếng Việt và dấu câu cơ bản
    text = re.sub(r'[^\w\s.,!?àáạảãâầấậẩẫăằắặẳẵèéẹẻẽêềếệểễìíịỉĩòóọỏõôồốộổỗơờớợởỡùúụủũưừứựửữỳýỵỷỹđ]', ' ', text)
    
    # 7. Loại bỏ khoảng trắng thừa
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text

print("✓ Đã định nghĩa hàm preprocess_text_basic()")

# ===== 3. TEST HÀM TIỀN XỬ LÝ =====
print("\n[3/6] Test hàm tiền xử lý với vài ví dụ:")
print("-" * 60)

test_samples = dataset['train']['sentence'][:3]
for i, text in enumerate(test_samples, 1):
    cleaned = preprocess_text_basic(text)
    print(f"\nMẫu {i}:")
    print(f"  Gốc:    {text}")
    print(f"  Cleaned: {cleaned}")

# ===== 4. ÁP DỤNG TIỀN XỬ LÝ CHO TOÀN BỘ DATASET =====
print("\n[4/6] Áp dụng tiền xử lý cho toàn bộ dataset...")
print("⏳ Đang xử lý... (có thể mất 10-30 giây)")

# Áp dụng cho tất cả các split
dataset = dataset.map(
    lambda x: {'cleaned_text': preprocess_text_basic(x['sentence'])},
    desc="Cleaning text"
)

print("✓ Đã làm sạch văn bản!")

# ===== 5. KIỂM TRA KẾT QUẢ =====
print("\n[5/6] Kiểm tra kết quả sau tiền xử lý:")
print("-" * 60)

# Chuyển sang DataFrame để dễ xem
train_df = pd.DataFrame(dataset['train'])

# Kiểm tra độ dài trước và sau
train_df['original_length'] = train_df['sentence'].apply(len)
train_df['cleaned_length'] = train_df['cleaned_text'].apply(len)
train_df['length_diff'] = train_df['original_length'] - train_df['cleaned_length']

print(f"\nThống kê độ dài:")
print(f"  Độ dài gốc TB:    {train_df['original_length'].mean():.1f} ký tự")
print(f"  Độ dài cleaned TB: {train_df['cleaned_length'].mean():.1f} ký tự")
print(f"  Giảm TB:          {train_df['length_diff'].mean():.1f} ký tự ({train_df['length_diff'].mean()/train_df['original_length'].mean()*100:.1f}%)")

# Kiểm tra xem có văn bản trống sau khi clean không
empty_texts = (train_df['cleaned_text'].str.strip() == '').sum()
if empty_texts > 0:
    print(f"\n⚠ Cảnh báo: Có {empty_texts} văn bản trống sau khi clean!")
    print("  → Nên xem xét loại bỏ hoặc giữ nguyên văn bản gốc")
else:
    print("\n✓ Không có văn bản trống sau khi clean!")

# Xem một vài ví dụ chi tiết
print("\n📝 So sánh 5 mẫu trước và sau:")
print("=" * 60)
for i in range(5):
    print(f"\nMẫu {i+1} - Sentiment: {train_df.iloc[i]['sentiment']}")
    print(f"  Gốc:    {train_df.iloc[i]['sentence'][:100]}...")
    print(f"  Cleaned: {train_df.iloc[i]['cleaned_text'][:100]}...")

# ===== 6. LƯU DATASET ĐÃ TIỀN XỬ LÝ =====
print("\n[6/6] Lưu dataset đã tiền xử lý...")

# Lưu dạng Hugging Face dataset
save_path = "./vietnamese_feedback_cleaned"
dataset.save_to_disk(save_path)
print(f"✓ Đã lưu vào '{save_path}'")

# Lưu mẫu CSV để kiểm tra
sample_df = train_df[['sentence', 'cleaned_text', 'sentiment', 'topic']].head(100)
sample_df.to_csv('cleaned_sample.csv', index=False, encoding='utf-8-sig')
print("✓ Đã lưu 100 mẫu vào 'cleaned_sample.csv'")

# ===== 7. THỐNG KÊ TỔNG KẾT =====
print("\n" + "=" * 60)
print("📊 THỐNG KÊ TỔNG KẾT:")
print("=" * 60)

stats_df = pd.DataFrame({
    'Split': ['Train', 'Validation', 'Test', 'TOTAL'],
    'Số mẫu': [
        len(dataset['train']),
        len(dataset['validation']),
        len(dataset['test']),
        len(dataset['train']) + len(dataset['validation']) + len(dataset['test'])
    ],
    'Độ dài TB (ký tự)': [
        pd.DataFrame(dataset['train'])['cleaned_text'].apply(len).mean(),
        pd.DataFrame(dataset['validation'])['cleaned_text'].apply(len).mean(),
        pd.DataFrame(dataset['test'])['cleaned_text'].apply(len).mean(),
        '-'
    ],
    'Số từ TB': [
        pd.DataFrame(dataset['train'])['cleaned_text'].apply(lambda x: len(x.split())).mean(),
        pd.DataFrame(dataset['validation'])['cleaned_text'].apply(lambda x: len(x.split())).mean(),
        pd.DataFrame(dataset['test'])['cleaned_text'].apply(lambda x: len(x.split())).mean(),
        '-'
    ]
})

print(stats_df.to_string(index=False))

print("\n" + "=" * 60)
print("✅ HOÀN THÀNH BƯỚC 2: TIỀN XỬ LÝ DỮ LIỆU")
print("=" * 60)

print("\n💡 NỘI DUNG ĐÃ LÀM:")
print("   ✓ Chuyển về chữ thường")
print("   ✓ Loại bỏ URL, email, HTML tags")
print("   ✓ Loại bỏ ký tự đặc biệt (giữ tiếng Việt)")
print("   ✓ Loại bỏ khoảng trắng thừa")
print("   ✓ Lưu dataset đã clean")

print("\n📁 FILE ĐÃ TẠO:")
print(f"   • {save_path}/ - Dataset đã clean (Arrow format)")
print("   • cleaned_sample.csv - 100 mẫu để kiểm tra")

```

```python
from datasets import load_from_disk
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import (
    accuracy_score, 
    precision_recall_fscore_support,
    classification_report,
    confusion_matrix
)
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
import time
import warnings
warnings.filterwarnings('ignore')

print("="*70)
print("BƯỚC 3A: TRAINING VỚI MACHINE LEARNING")
print("="*70)

# ===== 1. LOAD DATASET ĐÃ TIỀN XỬ LÝ =====
print("\n[1/8] Đang load dataset đã tiền xử lý...")

try:
    dataset = load_from_disk("./vietnamese_feedback_cleaned")
    print("✓ Load dataset đã clean thành công!")
except:
    print("⚠ Không tìm thấy dataset đã clean, load dataset gốc...")
    from datasets import load_dataset
    dataset = load_dataset("uitnlp/vietnamese_students_feedback")
    # Sử dụng cột 'sentence' gốc
    text_column = 'sentence'
else:
    text_column = 'cleaned_text'

# Chuyển sang DataFrame
train_df = pd.DataFrame(dataset['train'])
val_df = pd.DataFrame(dataset['validation'])
test_df = pd.DataFrame(dataset['test'])

print(f"✓ Train: {len(train_df):,} mẫu")
print(f"✓ Validation: {len(val_df):,} mẫu")
print(f"✓ Test: {len(test_df):,} mẫu")
print(f"✓ Sử dụng cột: '{text_column}'")

# ===== 2. CHUẨN BỊ DỮ LIỆU =====
print("\n[2/8] Chuẩn bị dữ liệu...")

X_train_text = train_df[text_column].values
X_val_text = val_df[text_column].values
X_test_text = test_df[text_column].values

y_train = train_df['sentiment'].values
y_val = val_df['sentiment'].values
y_test = test_df['sentiment'].values

label_names = {0: 'Negative', 1: 'Neutral', 2: 'Positive'}

print(f"✓ Đã chuẩn bị dữ liệu")
print(f"  Phân bố train: {dict(pd.Series(y_train).value_counts().sort_index())}")

# ===== 3. TẠO TF-IDF FEATURES =====
print("\n[3/8] Tạo TF-IDF features...")
print("⏳ Đang vectorize... (mất ~10-20 giây)")

vectorizer = TfidfVectorizer(
    max_features=5000,      # Giữ 5000 từ phổ biến nhất
    ngram_range=(1, 2),     # Unigram và bigram
    min_df=2,               # Từ phải xuất hiện ít nhất 2 lần
    max_df=0.8,             # Loại bỏ từ xuất hiện >80% văn bản
    sublinear_tf=True       # Scale logarithm cho TF
)

start_time = time.time()
X_train = vectorizer.fit_transform(X_train_text)
X_val = vectorizer.transform(X_val_text)
X_test = vectorizer.transform(X_test_text)
vectorize_time = time.time() - start_time

print(f"✓ Vectorize xong trong {vectorize_time:.2f} giây!")
print(f"  Shape: {X_train.shape}")
print(f"  Số features: {X_train.shape[1]:,}")
print(f"  Sparsity: {(1 - X_train.nnz / (X_train.shape[0] * X_train.shape[1])) * 100:.2f}%")

# ===== 4. ĐỊNH NGHĨA MODELS =====
print("\n[4/8] Định nghĩa models...")

models = {
    'Logistic Regression': LogisticRegression(
        max_iter=1000,
        random_state=42,
        class_weight='balanced',  # Xử lý class imbalance
        C=1.0
    ),
    'Linear SVM': LinearSVC(
        random_state=42,
        max_iter=2000,
        class_weight='balanced',
        C=1.0
    ),
    'Naive Bayes': MultinomialNB(
        alpha=0.1  # Laplace smoothing
    )
}

print(f"✓ Đã chuẩn bị {len(models)} models")

# ===== 5. TRAINING =====
print("\n[5/8] Bắt đầu training...")
print("-" * 70)

results = {}

for name, model in models.items():
    print(f"\n🔄 Training {name}...")
    
    # Train
    start_time = time.time()
    model.fit(X_train, y_train)
    train_time = time.time() - start_time
    
    # Predict
    y_pred_train = model.predict(X_train)
    y_pred_val = model.predict(X_val)
    y_pred_test = model.predict(X_test)
    
    # Evaluate
    train_acc = accuracy_score(y_train, y_pred_train)
    val_acc = accuracy_score(y_val, y_pred_val)
    test_acc = accuracy_score(y_test, y_pred_test)
    
    # Metrics chi tiết
    precision, recall, f1, _ = precision_recall_fscore_support(
        y_test, y_pred_test, average='weighted', zero_division=0
    )
    
    results[name] = {
        'model': model,
        'train_time': train_time,
        'train_acc': train_acc,
        'val_acc': val_acc,
        'test_acc': test_acc,
        'precision': precision,
        'recall': recall,
        'f1': f1,
        'y_pred_test': y_pred_test
    }
    
    print(f"  ⏱ Train time:     {train_time:.2f}s")
    print(f"  📊 Train Acc:     {train_acc:.4f} ({train_acc*100:.2f}%)")
    print(f"  📊 Val Acc:       {val_acc:.4f} ({val_acc*100:.2f}%)")
    print(f"  📊 Test Acc:      {test_acc:.4f} ({test_acc*100:.2f}%)")
    print(f"  📊 F1 Score:      {f1:.4f}")

# ===== 6. CHỌN MODEL TỐT NHẤT =====
print("\n" + "="*70)
print("[6/8] Chọn model tốt nhất...")
print("-" * 70)

best_model_name = max(results, key=lambda x: results[x]['test_acc'])
best_result = results[best_model_name]
best_model = best_result['model']

print(f"\n🏆 MODEL TỐT NHẤT: {best_model_name}")
print(f"  ✓ Test Accuracy:  {best_result['test_acc']:.4f} ({best_result['test_acc']*100:.2f}%)")
print(f"  ✓ F1 Score:       {best_result['f1']:.4f}")
print(f"  ✓ Precision:      {best_result['precision']:.4f}")
print(f"  ✓ Recall:         {best_result['recall']:.4f}")
print(f"  ✓ Train time:     {best_result['train_time']:.2f}s")

# ===== 7. CLASSIFICATION REPORT =====
print("\n[7/8] Classification Report chi tiết:")
print("="*70)
print(classification_report(
    y_test, 
    best_result['y_pred_test'],
    target_names=[label_names[i] for i in sorted(label_names.keys())],
    digits=4
))

# ===== 8. VISUALIZATIONS =====
print("\n[8/8] Tạo visualizations...")

fig = plt.figure(figsize=(16, 10))

# 1. So sánh accuracy các models
ax1 = plt.subplot(2, 3, 1)
model_names = list(results.keys())
train_accs = [results[m]['train_acc'] for m in model_names]
val_accs = [results[m]['val_acc'] for m in model_names]
test_accs = [results[m]['test_acc'] for m in model_names]

x = np.arange(len(model_names))
width = 0.25

ax1.bar(x - width, train_accs, width, label='Train', color='skyblue')
ax1.bar(x, val_accs, width, label='Validation', color='lightcoral')
ax1.bar(x + width, test_accs, width, label='Test', color='lightgreen')

ax1.set_ylabel('Accuracy', fontweight='bold')
ax1.set_title('So sánh Accuracy các Models', fontweight='bold', fontsize=14)
ax1.set_xticks(x)
ax1.set_xticklabels(model_names, rotation=15, ha='right')
ax1.legend()
ax1.grid(axis='y', alpha=0.3)
ax1.set_ylim([0.5, 1.0])

# 2. Confusion Matrix
ax2 = plt.subplot(2, 3, 2)
cm = confusion_matrix(y_test, best_result['y_pred_test'])
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax2,
            xticklabels=[label_names[i] for i in sorted(label_names.keys())],
            yticklabels=[label_names[i] for i in sorted(label_names.keys())],
            cbar_kws={'label': 'Count'})
ax2.set_title(f'Confusion Matrix - {best_model_name}', fontweight='bold', fontsize=14)
ax2.set_ylabel('True Label', fontweight='bold')
ax2.set_xlabel('Predicted Label', fontweight='bold')

# 3. Precision, Recall, F1 per class
ax3 = plt.subplot(2, 3, 3)
precisions, recalls, f1s, supports = precision_recall_fscore_support(
    y_test, best_result['y_pred_test'], labels=[0, 1, 2], zero_division=0
)

x_pos = np.arange(3)
width = 0.25

ax3.bar(x_pos - width, precisions, width, label='Precision', color='#3498db')
ax3.bar(x_pos, recalls, width, label='Recall', color='#e74c3c')
ax3.bar(x_pos + width, f1s, width, label='F1', color='#2ecc71')

ax3.set_ylabel('Score', fontweight='bold')
ax3.set_title('Metrics per Class', fontweight='bold', fontsize=14)
ax3.set_xticks(x_pos)
ax3.set_xticklabels([label_names[i] for i in sorted(label_names.keys())])
ax3.legend()
ax3.grid(axis='y', alpha=0.3)
ax3.set_ylim([0, 1.1])

# 4. Training time comparison
ax4 = plt.subplot(2, 3, 4)
train_times = [results[m]['train_time'] for m in model_names]
colors_time = ['#3498db', '#e74c3c', '#2ecc71']
bars = ax4.barh(model_names, train_times, color=colors_time)
ax4.set_xlabel('Time (seconds)', fontweight='bold')
ax4.set_title('Training Time Comparison', fontweight='bold', fontsize=14)
ax4.grid(axis='x', alpha=0.3)
for i, (bar, time_val) in enumerate(zip(bars, train_times)):
    ax4.text(time_val + 0.01, i, f'{time_val:.2f}s', va='center', fontweight='bold')

# 5. F1 Score comparison
ax5 = plt.subplot(2, 3, 5)
f1_scores = [results[m]['f1'] for m in model_names]
bars = ax5.bar(model_names, f1_scores, color=['#3498db', '#e74c3c', '#2ecc71'])
ax5.set_ylabel('F1 Score', fontweight='bold')
ax5.set_title('F1 Score Comparison', fontweight='bold', fontsize=14)
ax5.set_xticklabels(model_names, rotation=15, ha='right')
ax5.grid(axis='y', alpha=0.3)
ax5.set_ylim([0, 1.0])
for bar, score in zip(bars, f1_scores):
    height = bar.get_height()
    ax5.text(bar.get_x() + bar.get_width()/2., height + 0.01,
             f'{score:.4f}', ha='center', va='bottom', fontweight='bold')

# 6. Model summary table
ax6 = plt.subplot(2, 3, 6)
ax6.axis('off')

summary_data = []
for name in model_names:
    summary_data.append([
        name,
        f"{results[name]['test_acc']:.4f}",
        f"{results[name]['f1']:.4f}",
        f"{results[name]['train_time']:.2f}s"
    ])

table = ax6.table(cellText=summary_data,
                  colLabels=['Model', 'Test Acc', 'F1', 'Time'],
                  cellLoc='center',
                  loc='center',
                  bbox=[0, 0, 1, 1])
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1, 2)

# Highlight best model
for i, name in enumerate(model_names):
    if name == best_model_name:
        for j in range(4):
            table[(i+1, j)].set_facecolor('#90EE90')

ax6.set_title('Model Performance Summary', fontweight='bold', fontsize=14, pad=20)

plt.tight_layout()
plt.savefig('ml_training_results.png', dpi=300, bbox_inches='tight')
print("✓ Đã lưu biểu đồ vào 'ml_training_results.png'")
plt.show()

# ===== 9. LƯU MODEL =====
print("\n" + "="*70)
print("💾 Lưu model và vectorizer...")

joblib.dump(best_model, 'best_sentiment_model.pkl')
joblib.dump(vectorizer, 'tfidf_vectorizer.pkl')

print(f"✓ Đã lưu model vào 'best_sentiment_model.pkl'")
print(f"✓ Đã lưu vectorizer vào 'tfidf_vectorizer.pkl'")

# Lưu metadata
metadata = {
    'model_name': best_model_name,
    'test_accuracy': float(best_result['test_acc']),
    'f1_score': float(best_result['f1']),
    'precision': float(best_result['precision']),
    'recall': float(best_result['recall']),
    'train_time': float(best_result['train_time']),
    'num_features': X_train.shape[1],
    'label_names': label_names
}

import json
with open('model_metadata.json', 'w', encoding='utf-8') as f:
    json.dump(metadata, f, indent=2, ensure_ascii=False)
print("✓ Đã lưu metadata vào 'model_metadata.json'")

# ===== TỔNG KẾT =====
print("\n" + "="*70)
print("✅ HOÀN THÀNH BƯỚC 3A: TRAINING")
print("="*70)

print(f"\n🏆 KẾT QUẢ CUỐI CÙNG:")
print(f"   Model: {best_model_name}")
print(f"   Test Accuracy: {best_result['test_acc']*100:.2f}%")
print(f"   F1 Score: {best_result['f1']:.4f}")
print(f"   Training time: {best_result['train_time']:.2f}s")

print(f"\n📁 FILE ĐÃ TẠO:")
print(f"   • best_sentiment_model.pkl - Model đã train")
print(f"   • tfidf_vectorizer.pkl - TF-IDF vectorizer")
print(f"   • model_metadata.json - Metadata")
print(f"   • ml_training_results.png - Biểu đồ kết quả")

```

Dự Đoán 

```python
import joblib
import json
import numpy as np
import pandas as pd
import re
import warnings
warnings.filterwarnings('ignore')

print("="*70)
print("BƯỚC 4: DỰ ĐOÁN VÀ KIỂM THỬ MODEL")
print("="*70)

# ===== 1. LOAD MODEL VÀ VECTORIZER =====
print("\n[1/5] Load model và vectorizer...")

try:
    model = joblib.load('best_sentiment_model.pkl')
    vectorizer = joblib.load('tfidf_vectorizer.pkl')
    
    with open('model_metadata.json', 'r', encoding='utf-8') as f:
        metadata = json.load(f)
    
    print("✓ Load thành công!")
    print(f"  Model: {metadata['model_name']}")
    print(f"  Test Accuracy: {metadata['test_accuracy']*100:.2f}%")
    print(f"  F1 Score: {metadata['f1_score']:.4f}")
    
except FileNotFoundError as e:
    print(f"❌ Lỗi: Không tìm thấy file model!")
    print(f"   Hãy chạy BƯỚC 3 trước để train model")
    raise e

# Label mapping
label_names = {0: 'Tiêu cực 😞', 1: 'Trung lập 😐', 2: 'Tích cực 😊'}
label_names_simple = {0: 'Tiêu cực', 1: 'Trung lập', 2: 'Tích cực'}

# ===== 2. HÀM TIỀN XỬ LÝ (GIỐNG BƯỚC 2) =====
print("\n[2/5] Chuẩn bị hàm tiền xử lý...")

def preprocess_text(text):
    """Tiền xử lý văn bản giống BƯỚC 2"""
    if not isinstance(text, str):
        return ""
    
    text = text.lower()
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'<.*?>', '', text)
    text = re.sub(r'[^\w\s.,!?àáạảãâầấậẩẫăằắặẳẵèéẹẻẽêềếệểễìíịỉĩòóọỏõôồốộổỗơờớợởỡùúụủũưừứựửữỳýỵỷỹđ]', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text

print("✓ Đã chuẩn bị hàm tiền xử lý")

# ===== 3. HÀM DỰ ĐOÁN =====
print("\n[3/5] Tạo hàm dự đoán...")

def predict_sentiment(text, return_proba=True):
    """
    Dự đoán cảm xúc của văn bản
    
    Args:
        text (str): Văn bản cần dự đoán
        return_proba (bool): Có trả về xác suất không
    
    Returns:
        dict: Kết quả dự đoán
    """
    # Tiền xử lý
    cleaned_text = preprocess_text(text)
    
    # Vectorize
    X = vectorizer.transform([cleaned_text])
    
    # Predict
    prediction = model.predict(X)[0]
    
    result = {
        'text': text,
        'cleaned_text': cleaned_text,
        'label': int(prediction),
        'sentiment': label_names_simple[prediction],
        'sentiment_emoji': label_names[prediction]
    }
    
    # Thêm probability nếu model hỗ trợ
    if return_proba and hasattr(model, 'predict_proba'):
        probabilities = model.predict_proba(X)[0]
        result['confidence'] = float(max(probabilities))
        result['probabilities'] = {
            label_names_simple[i]: float(prob) 
            for i, prob in enumerate(probabilities)
        }
    elif return_proba and hasattr(model, 'decision_function'):
        # Với SVM dùng decision function
        scores = model.decision_function(X)[0]
        result['confidence'] = float(max(np.abs(scores)))
        result['decision_scores'] = {
            label_names_simple[i]: float(score) 
            for i, score in enumerate(scores)
        }
    
    return result

def predict_batch(texts):
    """Dự đoán cho nhiều văn bản cùng lúc"""
    results = []
    for text in texts:
        results.append(predict_sentiment(text))
    return results

print("✓ Đã tạo hàm predict_sentiment() và predict_batch()")

# ===== 4. TEST VỚI VÍ DỤ MẪU =====
print("\n[4/5] Test với các ví dụ mẫu:")
print("="*70)

test_sentences = [
    "Giảng viên dạy rất hay và nhiệt tình, tôi học được nhiều kiến thức bổ ích",
    "Môn học này quá khó và nhàm chán, tôi không hiểu gì cả",
    "Lớp học bình thường, không có gì đặc biệt",
    "Thầy giáo rất tâm huyết và chu đáo, luôn giúp đỡ sinh viên",
    "Nội dung quá lý thuyết, thiếu thực hành, không hứng thú",
    "Môn học khá hay nhưng hơi khó",
    "Giảng viên không nhiệt tình, giảng bài không rõ ràng",
    "Tôi rất thích môn học này, sẽ tiếp tục học các môn liên quan"
]

print("\n📝 Kết quả dự đoán:\n")

for i, text in enumerate(test_sentences, 1):
    result = predict_sentiment(text)
    
    print(f"{i}. Câu: \"{text}\"")
    print(f"   └─ Cảm xúc: {result['sentiment_emoji']}")
    
    if 'probabilities' in result:
        print(f"   └─ Độ tin cậy: {result['confidence']*100:.1f}%")
        probs_str = ", ".join([f"{k}: {v*100:.1f}%" for k, v in result['probabilities'].items()])
        print(f"   └─ Chi tiết: {probs_str}")
    elif 'decision_scores' in result:
        print(f"   └─ Độ tin cậy: {result['confidence']:.2f}")
    print()

# ===== 5. CHẾ ĐỘ TƯƠNG TÁC =====
print("\n[5/5] Chế độ tương tác:")
print("="*70)
print("💡 Nhập câu tiếng Việt để dự đoán cảm xúc")
print("💡 Nhấn Enter (để trống) để thoát")
print("💡 Gõ 'test' để xem thêm ví dụ")
print("💡 Gõ 'stats' để xem thống kê model")
print("-"*70)

def interactive_mode():
    """Chế độ tương tác"""
    while True:
        print()
        user_input = input("📝 Nhập câu: ").strip()
        
        if not user_input:
            print("\n👋 Tạm biệt!")
            break
        
        if user_input.lower() == 'test':
            # Random test
            import random
            random_tests = [
                "Giảng viên tốt quá",
                "Môn học tệ",
                "Bình thường thôi",
                "Rất hài lòng với khóa học",
                "Không muốn học nữa"
            ]
            print("\n🎲 Random tests:")
            for t in random_tests:
                r = predict_sentiment(t)
                print(f"  '{t}' → {r['sentiment_emoji']}")
            continue
        
        if user_input.lower() == 'stats':
            print("\n📊 Thống kê Model:")
            print(f"  • Model: {metadata['model_name']}")
            print(f"  • Test Accuracy: {metadata['test_accuracy']*100:.2f}%")
            print(f"  • F1 Score: {metadata['f1_score']:.4f}")
            print(f"  • Precision: {metadata['precision']:.4f}")
            print(f"  • Recall: {metadata['recall']:.4f}")
            continue
        
        # Dự đoán
        result = predict_sentiment(user_input)
        
        print(f"\n   🎯 Kết quả:")
        print(f"   ├─ Cảm xúc: {result['sentiment_emoji']}")
        
        if 'probabilities' in result:
            print(f"   ├─ Độ tin cậy: {result['confidence']*100:.1f}%")
            print(f"   └─ Phân bố:")
            for sent, prob in result['probabilities'].items():
                bar_length = int(prob * 30)
                bar = '█' * bar_length + '░' * (30 - bar_length)
                print(f"      {sent:>10}: [{bar}] {prob*100:.1f}%")
        elif 'decision_scores' in result:
            print(f"   └─ Độ tin cậy: {result['confidence']:.2f}")

# ===== 6. LƯU KẾT QUẢ TEST =====
print("\n💾 Lưu kết quả test...")

test_results = []
for text in test_sentences:
    result = predict_sentiment(text)
    test_results.append({
        'text': result['text'],
        'sentiment': result['sentiment'],
        'label': result['label']
    })

test_df = pd.DataFrame(test_results)
test_df.to_csv('test_predictions.csv', index=False, encoding='utf-8-sig')
print("✓ Đã lưu kết quả vào 'test_predictions.csv'")

# ===== TỔNG KẾT =====
print("\n" + "="*70)
print("✅ HOÀN THÀNH BƯỚC 4: DỰ ĐOÁN VÀ KIỂM THỬ")
print("="*70)

print("\n📦 CÁC HÀM CÓ SẴN:")
print("   • predict_sentiment(text) - Dự đoán 1 câu")
print("   • predict_batch(texts) - Dự đoán nhiều câu")
print("   • interactive_mode() - Chế độ tương tác")

print("\n📁 FILE ĐÃ TẠO:")
print("   • test_predictions.csv - Kết quả test")

print("\n🎮 THỬ NGAY:")
print("   Gọi: interactive_mode()")
print("   Hoặc: predict_sentiment('câu của bạn')")

print("\n" + "="*70)
print("🎉 HOÀN THÀNH TẤT CẢ CÁC BƯỚC!")
print("="*70)

print("\n💡 TỔNG KẾT DỰ ÁN:")
print("   ✅ BƯỚC 1: Khám phá dữ liệu (EDA)")
print("   ✅ BƯỚC 2: Tiền xử lý dữ liệu")
print("   ✅ BƯỚC 3: Training model ML")
print("   ✅ BƯỚC 4: Dự đoán và kiểm thử")

print(f"\n🏆 KẾT QUẢ CUỐI CÙNG:")
print(f"   Model: {metadata['model_name']}")
print(f"   Accuracy: {metadata['test_accuracy']*100:.2f}%")
print(f"   F1 Score: {metadata['f1_score']:.4f}")

```
